# Plot the following stacked bar chart

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 

countries = ['USA', 'GB', 'China', 'Russia', 'Germany']
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 18, 10])
golds = np.array([46, 27, 26, 19, 17])
ind = [x for x, _ in enumerate(countries)]

total = bronzes + silvers + golds

proportion_bronzes = np.divide(bronzes, total) * 100
proportion_silvers = np.divide(silvers, total) * 100
proportion_golds = np.divide(golds, total) * 100

plt.bar(ind, proportion_golds, width=0.8, label='golds', color='gold', bottom=proportion_bronzes+proportion_silvers)
plt.bar(ind, proportion_silvers, width=0.8, label='silvers', color='silver', bottom=proportion_bronzes)
plt.bar(ind, proportion_bronzes, width=0.8, label='bronzes', color='#CD853F')

plt.xticks(ind, countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.title("2012 Olympics Top Scorers' Medals by Proportion")
plt.ylim=1.0

# rotate axis labels
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')

plt.show()
```

# Step 1: Crawl stock web 

```python
%matplotlib inline
import pandas as pd

#df_list = pd.read_html('https://stock.wearn.com/cdata.asp?kind=2330')
df_list = pd.read_html('./dataset/wearn_stock_price/index.html')

df = df_list[0]
print(df_list[0])
```

# Step 2: Remove first two row and set new column name

```python
df.columns = ['time', 'open', 'high', 'low', 'close', 'volume']
df = df.drop([0,1], axis=0)
df = df.reset_index(drop=True)
print(df.head())
```

# Step 3: Change dataframe to time series 

```python
df['year'] = df['time'].apply(lambda x: str(int(x.split('/')[0])+1911))
df['date'] = df['time'].apply(lambda x: ''.join(x.split('/')[1:]))
df['new_time'] = df['year'] + df['date']

df['new_time'] = pd.to_datetime(df['new_time'])

# set pandas frame index as date
df.index = df['new_time']
df = df.drop(['time','year', 'date', 'new_time'], axis=1)
print(df.head())
```

# Step 4: Change stock price to numerical  
## hint: use df[index_name].astype('float64')

```python
df.dtypes
```

```python
df['open'] = df['open'].astype('float64')
df['high'] = df['high'].astype('float64')
df['low'] = df['low'].astype('float64')
df['close'] = df['close'].astype('float64')
df['volume'] = df['volume'].astype('float64')
df.dtypes
```

# Step 5: Plot time versus close price
